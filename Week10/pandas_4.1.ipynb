```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_excel('  .xlsx', engine='openpyxl', header=0)

df = df.fillna(method='ffill')

mask = (df[''] == '') & (df[''] != '')
df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis=1)
df_seoul.rename({'': ''}, axis =1, inplace = True)
df_seoul.set_index('', inplace=True)
sr_one = df_seoul.loc['']
plt.plot(sr_one.index, sr_one.values)
plt.plot(sr_one)

sr_one = df_seoul.loc['']
plt.plot(sr_one.index, sr_one.values)
plt.title('->   ')
plt.xlabel('')
plt.ylabel(' ')
plt.show()

sr_one = df_seoul.loc['']
plt.figure(figsize=(14, 5))
plt.xticks(rotation='vertical')
plt.plot(sr_one.index, sr_one.values)
plt.title(' ->   ')
plt.xlabel('')
plt.ylabel(' ')
plt.legend(labels=[' -> '], loc='best')
plt.show()

sr_one = df_seoul.loc['']
plt.style.use('ggplot')
plt.figure(figsize=(14, 5))
plt.xticks(size=10, rotation='vertical')
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10)
plt.title(' ->   ', size=30)
plt.xlabel('', size=20)
plt.ylabel(' ',size=20)
plt.legend(labels=[' -> '], loc='best', fontsize=15)
plt.show()

plt.annotate('',
             xy=(20, 620000),       #  ()
             xytext=(2, 290000),    #  ()
             xycoords='data',       #
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), # 
             )

plt.annotate('',
             xy=(47, 450000),       #  ()
             xytext=(30, 580000),   #  ()
             xycoords='data',       #
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  # 
             )

#   - 
plt.annotate(' (1970-1995)',  # 
             xy=(10, 550000),            #  
             rotation=25,                # 
             va='baseline',              #  
             ha='center',                #  
             fontsize=15,                # 
             )

plt.annotate(' (1995-2017)',  # 
             xy=(40, 560000),            #  
             rotation=-11,               # 
             va='baseline',              #  
             ha='center',                #  
             fontsize=15,                # 
             )

plt.show()

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(sr_one, 'o', markersize=10)
ax2.plot(sr_one, marker='o', markerfacecolor='green', markersize=10, color='olive', linewidth=2, label=' -> ')
ax2.legend(loc='best')

ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)
plt.show()

fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1)

ax.plot(sr_one, marker='o', markerfacecolor='green', markersize=10, color='olive', linewidth=2, label=' -> ')
ax.legend(loc='best')

ax.set_ylim(50000, 800000)
ax.set_title(' ->   ', size=20)
ax.set_xlabel('', size=12)
ax.set_ylabel(' ', size=12)

ax.set_xticklabels(sr_one.index, rotation=75)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()

col_years = list(map(str, range(1970, 2018)))
df_3 = df_seoul.loc[['', '', ''], col_years]
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, df_3.loc['',:], marker='o', markerfacecolor='green', markersize=10, color='olive', linewidth=2, label=' -> ')
ax.plot(col_years, df_3.loc['',:], marker='o', markerfacecolor='blue', markersize=10, color='skyblue', linewidth=2, label=' -> ')
ax.plot(col_years, df_3.loc['',:], marker='o', markerfacecolor='red', markersize=10, color='magenta', linewidth=2, label=' -> ')
ax.legend(loc='best')
ax.set_title(' -> , ,   ', size=20)

ax.set_xlabel('', size=12)
ax.set_ylabel(' ', size=12)
ax.set_xticklabels(col_years, rotation=90)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()

col_years = list(map(str, range(1970, 2018)))
df_4 = df_seoul.loc[['', '', '', ''], col_years]
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='green', markersize=10, color='olive', linewidth=2, label=' -> ')
ax2.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='blue', markersize=10, color='skyblue', linewidth=2, label=' -> ')
ax3.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='red', markersize=10, color='magenta', linewidth=2, label=' -> ')
ax4.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='orange', markersize=10, color='yellow', linewidth=2, label=' -> ')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_title(' ->   ', size=15)
ax2.set_title(' ->   ', size=15)
ax3.set_title(' ->   ', size=15)
ax4.set_title(' ->   ', size=15)

ax1.set_xticklabels(col_years, rotation=90)
ax2.set_xticklabels(col_years, rotation=90)
ax3.set_xticklabels(col_years, rotation=90)
ax4.set_xticklabels(col_years, rotation=90)
plt.show()
```
